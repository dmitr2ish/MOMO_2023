## класс DataFrame

двумерный массив с гибкими индексами строк и гибкими именами столбцов

Объект `DataFrame` можно рассматривать как упорядоченную последовательность объектов `Series` с одинаковым индексом

Метод для создания `DataFrame`:

`pd.DataFrame(data, columns=columns)`

```python
dataframe = pd.DataFrame(np.random.random((10,4)), columns=['a','b','c','d',])
dataframe
```

### Обращение к элементам

```python
dataframe.a
```

```python
dataframe['a']
```

• `loc` - явный индекс


```python
dataframe.loc[1]
```

```python
dataframe.loc[1:3]
```

```python
dataframe.loc[1:3, 'a']
```

```python
dataframe.loc[2:5, 'b':'d']
```

• `iloc` - неявный индекс

```python
dataframe.iloc[1]
```

```python
dataframe.iloc[1:3]
```

## Операция GroupBy

*  Шаг разбиения включает разделение на части и группировку объекта DataFrame на основе значений заданного ключа.
* Шаг применения включает вычисление какой-либо функции, обычно агрегирующей, преобразование или фильтрацию в пределах отдельных групп.
* На шаге объединения выполняется слияние результатов этих операций в выходной массив.

Добавим столбец с ключами

```python
dataframe['label'] = np.array(['A','A','B','C','A','B','A','C','B','C'])
dataframe
```

```python
dataframe.groupby('label')
```

```python
dataframe.groupby('label').sum()
```

### Индексация groupby

```python
dataframe.groupby('label')[['b','c']].sum()
```

## Агрегирование

Выполнение нескольких операций в рамках одной команды

```python
dataframe.groupby('label').aggregate(['min', max])
```

Для разных столбцов - своя агрегация

```python
dataframe.groupby('label').aggregate({'a': 'min',
                                      'b': 'max',
                                      'c': 'sum',
                                      'd': ['mean', 'std']
                                      })
```
